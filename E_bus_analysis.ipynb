## E-Bus Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data\E-Bus\Electric Bus Sales.csv")
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['financial_year'].str[:4] + '-' + df['month'], format='%Y-%b')

# Aggregate number of vehicles per month across all financial years
monthly_data = df.groupby('date')['num_vehicles'].sum()

plt.plot(monthly_data.index, monthly_data.values, linestyle='-')

plt.xlabel('Month')
plt.ylabel('Number of E-Buses')
plt.title('Number of Units Sold Over Time')

plt.xticks(rotation=45)

plt.grid(True)  
plt.tight_layout()  
plt.show()
```
