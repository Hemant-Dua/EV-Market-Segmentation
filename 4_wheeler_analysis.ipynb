## 4-Wheeler Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data\E-4wheelers\Electric 4-Wheeler Sales.csv")
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['financial_year'].str[:4] + '-' + df['month'], format='%Y-%b')

# Aggregate number of vehicles per month across all financial years
monthly_data = df.groupby('date')['num_vehicles'].sum()

plt.plot(monthly_data.index, monthly_data.values, linestyle='-')

plt.xlabel('Month')
plt.ylabel('Number of 4-Wheelers')
plt.title('Number of 4-Wheelers Over Time')

plt.xticks(rotation=45)

plt.grid(True)  
plt.tight_layout()  
plt.show()
```

## Customer Satisfaction Analysis (E-4Wheelers)

```python
df1 = pd.read_csv("data/E-4wheelers/4-wheeler-EV-cardekho.csv")
df2 = pd.read_csv("data/E-4wheelers/4-wheeler-EV-carwale.csv")
```

```python
df1.head()
```

```python
df1["Model"].unique()
```

```python
model = ['hyundai kona', 'Kia EV6', 'Tata Nexon EV', 'MG ZS EV',
       'Tata Tigor EV', 'BYD E6']

rating = []
for i in model:
    data = df1["Rating"].loc[df1["Model"] == i].mean()
    rating.append(data)

plt.bar(model, rating, width=0.5)

plt.xlabel('Model Name')
plt.ylabel('Average Rating')
plt.title('Average Rating per Model (Source: CarDekho.com)')

plt.xticks(rotation=90)
plt.show()
```
